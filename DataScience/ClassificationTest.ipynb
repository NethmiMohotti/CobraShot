```python
import numpy as np
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from keras.preprocessing import image
from keras.models import load_model
```

```python
IMAGE_SIZE = [224, 224]
```

```python
#trained model is loaded
cobrashot_m=load_model('cobrashot_m.h5')
```

```python
#fetch and load image from the location
dir_path="C:\\Users\\Nethmi Mohotti\\Desktop\\CobraShot\\Data Science\\dataset\\test"
for i in os.listdir(dir_path ):
    img = image.load_img(dir_path+"//"+i, target_size=(224, 224))
    plt.imshow(img)
    plt.show()
    x =image.img_to_array(img)
    x = np.expand_dims(x,axis=0)
    images = np.vstack([x]) 
    
    pred = cobrashot_m.predict(images)
    prediction= pred[0]
    for k in range(7):
        if prediction[k] == 1:
          print("Snake identified to the index")
        else:
          print('Unidentified')
```
